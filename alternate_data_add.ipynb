```python
# add in alternate external data for analysis
import pandas as pd
import os

current_datadir = './trimmed_data_and_plots/'
alternate_data = './alternate_data/'
```

```python

files = os.listdir(current_datadir)

produce_items = []
for n in files:
    if n.endswith('.csv'):
        produce_items.append(n)
```

```python
# import alt data oil:
oilprices = pd.read_csv(alternate_data+'POILWTIUSDM.csv', index_col=0)

# import stock data:
yfprices = pd.read_csv(alternate_data+'yftickers.csv', index_col=0)
```

```python
joined_alt = pd.concat([yfprices, oilprices], axis=1)
joined_alt
```

```python
joined_alt.index = pd.to_datetime(joined_alt.index)

for i in produce_items:
    producedata = pd.read_csv(current_datadir+i, index_col=0)

    producedata.index = pd.to_datetime(producedata.index)

    joined_data = pd.concat([producedata, joined_alt], axis=1)

    joined_data.to_csv(current_datadir+i)
```
