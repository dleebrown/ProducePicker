```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

plt.rcParams['xtick.labelsize']=14
plt.rcParams['ytick.labelsize']=14
```

```python
# pie chart of restaurant opex
expense_category = ['Rent', 'Food','Labor', 'Misc.', 'Profit']
expense_amount = [10, 30, 30, 20, 10]
explode = (0, 0.15, 0.0, 0, 0)

fig1, ax1 = plt.subplots(figsize=(5,5))
patches, texts= ax1.pie(expense_amount, explode=explode, labels=expense_category, shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
[ _.set_fontsize(14) for _ in texts ]


plt.savefig('restaurant_breakdown.pdf', format='pdf', bbox_inches='tight')

plt.show()

```
